## A Dynamic Statistical Comparison benchmark

We implemented the numerical study of the manuscript in [the DSC framework](https://stephenslab.github.io/dsc-wiki/). 
The benchmark is a single [SoS Notebook](https://vatlab.github.io/sos-docs/) file (a Jupyter notebook with SoS kernel, in JSON format) `finemapping.ipynb` that conslidates all scripts required for the benchmark.

To run the benchmark, first export scripts from the notebook:

```
./export.sos
```

Then run selected sub-benchmarks. For the manuscript we executed 2 benchmarks:

## Comparison with other methods on a subset of genotypes for cis-QTL analysis

```
dsc susie.dsc --target run_comparison -o susie_comparison --host susie_comparison.yml -c 60
```

## Comparison with DAP-G on full cis-QTL genotypes with 10 causal SNPs

```
dsc susie.dsc --target hard_case -o hard_case --host hard_case.yml -c 60
```

The configuration files `*.yml` configured the benchmark execution environment on a cluster system operated by the University of Chicago. It is also possible to run the benchmark on a local computer, if `--host` option is dropped. The `*.yml` files are still relevant in providing information on required resource for each DSC module. It is estimated that benchmark commands above will take about a week's computation on a single desktop machine with 28 CPU threads.

## Benchmark data download

The benchmark data can be downloaded [here](http://shiny.stephenslab.uchicago.edu), a XX GB `tar.gz` file including all intermediate results generated by the benchmark. This dataset can be used to reproduce numerical studies of the manuscript as [documented here](numerical_results.html).
